### RFM Analysis

```python
retail_df["InvoiceDate"].max()
```

```python
today_date=dt.datetime(2011,12,11)
```

```python
rfm=retail_df.groupby("CustomerID").agg({"InvoiceDate": lambda date:(today_date-date.max()).days,
                                      "InvoiceNo": lambda num:num.nunique(),
                                       "total_price": lambda totalprice:totalprice.sum()})
```

```python
rfm.columns=["Recency","Frequancy","Monetary"]
```

```python
rfm
```

```python
rfm=rfm[rfm["Monetary"]>0]
```

```python
rfm.describe().T
```

```python
rfm["Recency_score"]=pd.qcut(rfm["Recency"],5,labels=[5,4,3,2,1])
rfm["Frequancy_score"]= pd.qcut(rfm["Frequancy"].rank(method="first"),5,labels=[1,2,3,4,5])
rfm["Monetary_score"] = pd.qcut(rfm["Monetary"],5,labels =[1,2,3,4,5])
```

```python
rfm.sample(10)
```

```python
rfm["rfm_Score"] =( rfm["Recency_score"].astype(str) + rfm["Frequancy_score"].astype(str))
```

```python
rfm.sample(5)
```

```python
seg_map ={
    r'[1-2][1-2]' :'hbernating',
    r'[1-2][3-4]' : 'at_Resik',
    r'[1-2]5' : "cant_lose"   ,
    r'3[1-2]' : 'about_to_sleap',
    r'33': 'need_attention',
    r'[3-4][4-5]' : 'loyal_customer',
    r'41' : 'promising' ,
    r'51' : 'new_customers',
    r'[4-5][2-3]' : 'potianial_loyatlist',
    r'5[4-5]' : 'champions' 

}
rfm["segment"] =rfm["rfm_Score"].replace(seg_map,regex=True)
```

```python
rfm.sample(10)
```

```python
rfm[["segment","Recency","Frequancy","Monetary"]].groupby("segment").agg(['mean',"count","max"]).round()
```
